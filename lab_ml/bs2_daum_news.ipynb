*   다음 메인 페이지에서 검색어 입력하고 검색
*   검색 결과 페이지에서 뉴스 섹션을 클릭(선택)
*   뉴스 제목과 링크를 크롤링

```python
# 웹 서버에 요청(request)을 보내고 응답(response)을 받는 기능을 가지고 있는 모듈
import requests

# HTML 분석(parsing)
from bs4 import BeautifulSoup
```

```python
# 요청(request)를 보낼 주소(URL)
url1 = 'https://search.daum.net/search?w=news&nil_search=btn&DA=NTB&enc=utf8&cluster=y&cluster_page=1&q=%EC%84%A4+%EC%97%B0%ED%9C%B4'
```

```python
# 요청 주소(URL)로 GET 방식의 요청을 보내고 응답을 받음.
response = requests.get(url1)
print(response)  #> 200 - 응답 성공
```

```python
# 응답(response)에 포함된 HTML 코드(문자열)을 추출.
html = response.text.strip()
```

```python
print(html[:200])  # HTML 문자열에서 첫 200글자
```

```python
print(html[-200:])  # HTML 문자열에서 마지막 200글자
```

```python
# HTML 분석하는 BeautifulSoup 객체 생성
bs = BeautifulSoup(markup=html, features='html5lib')
```

```python
selector = 'ul.c-list-basic div.item-title a'
```

```python
elements = bs.select(selector)
```

```python
for x in elements:
    print(x.text.strip())
    print(x.get('href'))
```

뉴스 검색 결과를 1페이지부터 5페이지까지 출력

```python
for page in range(1, 6):
    print(f'\n**** page {page} *****')
    url = f'https://search.daum.net/search?w=news&nil_search=btn&DA=PGD&enc=utf8&cluster=y&cluster_page=1&q=%EC%84%A4+%EC%97%B0%ED%9C%B4&p={page}'
    response = requests.get(url)
    print(response)
    html = response.text.strip()
    bs = BeautifulSoup(markup=html, features='html5lib')
    selector1 = 'ul.c-list-basic div.item-title a'  # 기사 제목
    selector2 = 'ul.c-list-basic div.item-contents a'  # 기사 내용 일부
    elements1 = bs.select(selector1)
    elements2 = bs.select(selector2)
    for title, content in zip(elements1, elements2):
        print(title.get('href'))
        print(title.text)
        print(content.text)
```
