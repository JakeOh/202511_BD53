BeautifulSoup 라이브러리를 사용해서 Google Drive에 저장된 HTML 문서를 분석.

```python
from bs4 import BeautifulSoup
```

```python
web2_html = '/content/drive/MyDrive/Colab Notebooks/lab_ml/web2.html'
```

```python
# 1. HTML 문서를 오픈.
html = open(file=web2_html)  # mode='rt': 텍스트 읽기 모드. 기본값
```

```python
# 2. HTML 문석를 분석(parsing)할 수 있는 BeautifulSoup 객체 생성
soup = BeautifulSoup(markup=html, features='html5lib')
#> markup 파라미터: 오픈된 HTML 파일(스트림) 객체.
#> features 파라미터: HTML 문서를 분석(parsing)하는 라이브러리. 기본값은 'html.parser'.
```

```python
# 3. 생성된 BeautifulSoup 객체의 메서드를 사용해서 HTML 문서의 여러가지 요소들을 찾음.
# bs.find(name='태그이름'): HTML 문서에서 가장 처음 나오는 태그 요소를 찾아서 리턴.
element = soup.find(name='h1')
print(element)
```

```python
element = soup.find(name='a')
print(element)  #> HTML 문서의 a 태그 요소들 중에서 첫번째 a 요소만 리턴.
```

```python
print(element.text)  #> a 태그 요소의 컨텐트 텍스트
print(element.get(key='href'))  #> a 태그의 href 속성에 설정된 값
```

```python
# bs.find_all(name='태그이름'): 같은 태그를 갖는 모든 요소들의 리스트를 리턴.
links = soup.find_all(name='a')
# print(links)  #> find_all 메서드의 리턴 값은 리스트

for x in links:
    # print(x)
    print(x.text, x.get(key='href'))
```

```python
# 열려져 있는 파일 닫기
html.close()
```
