# NumPy

```python
import numpy as np
```

# `numpy.ndarray` 클래스 속성들(properties)

```python
# 2차원 배열(2-dimensional array)
arr = np.ones(shape=(2, 3))  #> 모든 원소가 1인 배열
```

```python
arr
```

```python
arr.shape  # shape 속성: 배열의 각 축(axis)을 따라서 있는 원소 개수들의 튜플(tuple)
```

```python
arr.size  # size 속성: 배열의 모든 원소 개수
```

```python
arr.ndim  # ndim 속성: 배열의 차원 개수. 축(axis)의 개수.
```

```python
arr.dtype  # dtype 속성: 배열 원소의 데이터 타입(data type)
```

```python
# 1차원 배열(1-dimensional array)
arr = np.arange(10)
print(arr)
print('shape:', arr.shape)
print('size:', arr.size)
print('ndim:', arr.ndim)
print('dtype:', arr.dtype)
```

```python
# 3차원 배열(3-dimensional array)
arr = np.arange(1, 25).reshape((2, 3, 4))
print(arr)
```

```python
print('shape:', arr.shape)
print('size:', arr.size)
print('ndim:', arr.ndim)
print('dtype:', arr.dtype)
```

# Indexing

인덱스를 사용해서 배열 원소를 참조.

```python
print(arr)  #> 3d array
```

```python
print(arr[0])  # 2d array
```

```python
print(arr[0][0])  # 1d array
```

```python
print(arr[0, 0])
```

```python
print(arr[0][0][0])  #> 값(int) --> scalar
```

```python
print(arr[0, 0, 0])
```

*   Python list 인덱스 사용 방법:
    *   `list[i][j][k]...`
*   NumPy ndarray 인덱스 사용 방법:
    *   `array[i][j][k]...`
    *   `array[i, j, k, ...]`
