# 배열 모양(shape) 변경

*   `reshape()`
*   `ravel()`

## reshape

```python
arr_1d = np.arange(12)
print(arr_1d)
```

```python
# arr_1d를 (4, 3) shape의 2d array로 변환
arr_2d = arr_1d.reshape((4, 3))
print(arr_2d)
```

```python
# reshape 메서드에 아규먼트를 전달할 때, 자동으로 계산할 수 있는 차원의 값은 -1을 사용할 수 있음.
arr_2d = arr_1d.reshape((4, -1))  # (4, 3)
print(arr_2d)
```

```python
arr_2d = arr_1d.reshape((-1, 4))  # (3, 4)
print(arr_2d)
```

```python
arr_1d = np.arange(24)
print(arr_1d)
```

```python
# arr_1d를 (2, 3, 4) 모양의 3d array로 변환
arr_3d = arr_1d.reshape((2, 3, 4))
print(arr_3d)
```

```python
arr_3d = arr_1d.reshape((2, 3, -1))
print(arr_3d)
```

```python
arr_3d = arr_1d.reshape((-1, 3, 4))
print(arr_3d)
```

## ravel vs flatten

*   ravel
    *   n차원 배열을 1차원 배열로 변환한 "뷰(view)"를 리턴.
    *   새로운 1차원 배열을 만드는 것이 아니라, 원본 배열의 모양을 1차원처럼 보여줄 뿐.
    *   리턴된 1차원 배열에서 원소를 변경하면, 원본 배열도 변경됨!
*   flatten
    *   n차원 배열을 1차원 배열로 변환한 "복사본(copy)"을 리턴.
    *   원본 배열과는 별개의 새로운 1차원 배열을 메모리에 만들어서 리턴.
    *   리턴된 1차원 배열의 원소를 변경하는 것은 원본 배열에 영향을 주지 않음!


```python
arr_2d = np.arange(12).reshape((3, 4))
print(arr_2d)
```

```python
flattened = arr_2d.flatten()  # 2d --> 1d
print(flattened)
```

```python
flattened[0] = 123
print(flattened)
```

```python
print(arr_2d)  # 원본 2d array은 변경되지 않음.
```

```python
raveled = arr_2d.ravel()  # 2d --> 1d
print(raveled)
```

```python
raveled[0] = 123
print(raveled)
```

```python
print(arr_2d)  # 원본 2d array이 변경됨!
```

# 배열 합치기(concatenate)

*   `np.concatenate([array1, array2, ...], axis=0)`: 특정 축(axis) 방향으로 배열 이어붙이기.
*   `np.r_[array1, array2]`: 행들을 이어붙이기. axis=0 방향으로 이어붙이기.
*   `np.c_[array1, array2]`: 컬럼들을 이어붙이기. axis=1 방향으로 이어붙이기.


```python
# (2, 3) shape의 2d array
arr1 = np.arange(6).reshape((2, 3))
print(arr1)
```

```python
# (1, 3) shape의 2d array
arr2 = np.array([[10, 20, 30]])
print(arr2)
```

```python
# (2, 1) shape의 2d array
arr3 = np.array([[11], [12]])
print(arr3)
```

```python
result = np.concatenate([arr1, arr2])  # axis=0(기본값) 생략 가능
print(result)
```

```python
result = np.concatenate([arr1, arr3], axis=1)
print(result)
```

```python
np.r_[arr1, arr2]
```

```python
np.c_[arr1, arr3]
```

```python
arr1 = np.arange(6).reshape((2, 3))
print(arr1)
```

```python
arr2 = np.array([10, 20, 30])
print(arr2)
```

```python
# np.concatenate([arr1, arr2], axis=0)
#> ValueError 발생. 차원이 다르면 이어붙이기가 되지 않음.
```

# Broadcasting

*   numpy.ndarray의 연산(+, -, *, /, ==, !=, >, <, ...)은 같은 인덱스(위치)의 원소들(element-wise)끼리 연산이 수행됨.
*   서로 모양(shape)이 다른 배열끼리도 element-wise 연산이 가능한 경우가 있음 --> broadcast
*   broadcasting이 가능한 경우
    *   누락되었거나 길이가 1인 차원에 대해서 broadcasting이 수행.
    *   이어지는 차원의 축의 길이가 같은 경우에 broadcasting이 가능.

## ndarray와 scalar의 연산

항상 연산이 가능(항상 브로드캐스트가 가능). 스칼라를 배열의 각 원소마다 연산 수행.

```python
a = np.array([1, 2, 3])
print(a)
print(a + 10)
```

## 2d array와 2d array의 broadcast

```python
a = np.arange(1, 7).reshape((2, 3))  # shape: (2, 3)
print(a)
```

```python
b = np.array([[7, 8, 9]])  # shape: (1, 3)
print(b)
```

```python
print(a + b)
```

```python
c = np.array([[11], [12]])  # shape: (2, 1)
print(c)
```

```python
print(a + c)
```
