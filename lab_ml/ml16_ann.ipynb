# 이진 분류(binary classification)

*   2개의 특성만을 사용
*   0(setosa)과 0이 아닌(versicolor, virginica) 클래스로 분류

```python
x = data[:, :2]
```

```python
x[:5]
```

```python
y = (target != 0).astype('int')
y
```

# SGD Classifier(확률적 경사하강법 분류기)

```python
# loss='log_loss': logistic 손실 함수를 최소화하는 확률적 경사 하강법. 확률 -> 분류.
# max_iter=10: epoch(모든 샘플을 사용해서 경사하강법을 수행) 횟수를 10번.
sgd = SGDClassifier(loss='log_loss', max_iter=10, n_jobs=-1, random_state=42)
```

```python
sgd.fit(X=x, y=y)  # 모델 훈련 -> 손실함수를 최소로 만드는 가중치(절편과 계수들) 찾음.
```

```python
sgd.coef_
```

```python
sgd.intercept_
```

$
z = -53 + 107 x_1 - 163 x_2
$

```python
# SGD의 예측 확률
y_prob = sgd.predict_proba(X=x)
```

```python
y_prob[:5]
```

```python
y_prob[-5:]
```

```python
# SGD 모델의 예측값
y_pred = sgd.predict(X=x)
y_pred
```

```python
# SGD 모델의 정확도
accuracy_score(y_true=y, y_pred=y_pred)
```

```python
149/150
```

# 신경망(Neural Network)

```python
# 입력층(input layer)
inputs = keras.Input(shape=(2,))
```

```python
# 출력층(output layer)
dense = keras.layers.Dense(units=1, activation='sigmoid')
```

```python
# 신경망 모델 생성
model = keras.Sequential(layers=[inputs, dense])
```

```python
# 생성된 신경망 모델 정보 요약
model.summary()
```

```python
# 신경망 모델 컴파일: optimizer(최적화 알고리즘), loss(손실함수), metrics(평가 기준)을 설정
model.compile(optimizer=keras.optimizers.SGD(),
              loss=keras.losses.binary_crossentropy,
              metrics=[keras.metrics.binary_accuracy])
```

```python
# 신경망 모델 훈련
model.fit(x=x, y=y, batch_size=1, epochs=10)
```

```python
# 신경망 모델이 학습을 통해서 찾은 모델 파라미터들(가중치, weights)
model.weights
```

```python
# 신경망 모델의 예측값
y_pred = model.predict(x=x)
y_pred.shape  #> (n_samples, 1): 150개 샘플들의 양성(1)일 예측 확률
```

```python
y_pred[:5]
```

```python
y_pred[-5:]
```

```python
y_pred_label = (y_pred > 0.5).astype('int').ravel()
y_pred_label
```

```python
# 신경망 모델 평가
model.evaluate(x=x, y=y)  #> [loss, accuracry]
```

# 은닉층(hidden layer)을 가지고 있는 신경망

```python
inputs = keras.layers.Input(shape=(2,))  # 입력층
dense1 = keras.layers.Dense(units=2, activation='sigmoid')  # 은닉층
dense2 = keras.layers.Dense(units=1, activation='sigmoid')  # 출력층
```

```python
model = keras.Sequential(layers=[inputs, dense1, dense2])  # 신경망
```

```python
model.summary()
```

```python
# 모델 컴파일: 어떤 손실함수를 어떤 평가 기준으로 어떻게 최적화할 것인가를 설정
model.compile(optimizer=keras.optimizers.SGD(),
              loss=keras.losses.binary_crossentropy,
              metrics=[keras.metrics.binary_accuracy])
```

```python
# 모델 훈련
model.fit(x=x, y=y, batch_size=30, epochs=10)
```

```python
model.weights
```

```python
model.evaluate(x=x, y=y)
```

# 신경망

4개 특성을 모두 사용한 다중 클래스(0/1/2) 분류

```python
model = keras.Sequential()  # Sequential(완전 연결 신경망) 객체 생성
```

```python
# 신경망에 입력층을 추가
model.add(keras.layers.Input(shape=(4,)))
```

```python
# 신경망에 출력층을 추가
model.add(keras.layers.Dense(units=3, activation='softmax'))
```

```python
model.summary()
#> # of params = 4 features x 3 units + 3 bias = 15
```

```python
# 모델 컴파일
model.compile(optimizer=keras.optimizers.SGD(),
              loss=keras.losses.sparse_categorical_crossentropy,
              metrics=[keras.metrics.sparse_categorical_accuracy])
```

```python
# 모델 훈련
model.fit(x=data, y=target, epochs=10)
```

```python
model.weights
#> kernel: (4, 3)
#> bias: (3,)
```

```python
predicts = model.predict(x=data)
```

```python
predicts.shape  #> (n_samples, n_classes)
```

```python
predicts[:5]
```

```python
model.evaluate(x=data, y=target)
```

```python
# epoch 횟수를 증가시켜서 훈련
model.fit(x=data, y=target, epochs=1_000, verbose=0)
```

```python
model.evaluate(x=data, y=target)
```
