## 계층적 인덱스(hierachical index), multi-level index

```python
s = pd.Series(data=np.random.randint(100, size=6), name='score',
              index=[['m', 'm', 'm', 'f', 'f', 'f'], ['a', 'b', 'c', 'a', 'b', 'c']])
s
```

```python
s.values  #> 1차원 배열(ndarray)
```

```python
s.index  #> MultiIndex - 튜플들을 아이템으로 갖는 배열
```

```python
s.index.nlevels  #> 계층(level)들의 개수
```

```python
s.iloc[-1]
```

### Multi-level index에서 indexing, slicing

nlevels(인덱스의 계층 개수)가 2개 이상인 multi-level 인덱스를 loc 속성에서 사용할 때,
*   첫번째 계층(level)의 인덱스만 가지고 indexing과 slicing을 할 수 있음.
*   두번째 계층의 인덱스만 가지고는 indexing/slicing을 할 수 없음!
*   튜플 형태의 인덱스로 indexing/slicing을 할 수 있음.


```python
s
```

```python
s.loc['m']  # 첫번째 계층(level)의 인덱스로  indexing.
```

```python
s.loc['f']
```

```python
s.loc['a']  #> KeyError - 'a'는 첫번째 계층(level)의 인덱스가 아니기 때문에.
```

```python
s.loc[('m', 'a')]  #> 튜플 형태의 인덱스를 사용한 indexing.
```

```python
s.loc[[('m', 'a'), ('f', 'a')]]
```

```python
s_idx_sorted = s.sort_index()
s_idx_sorted
```

```python
s_idx_sorted.loc['f':'m']
#> slicing은 인덱스들이 정렬된 상태일 때 가능.
#> 첫번째 계층(level)의 인덱스만으로도 slicing 가능.
```

```python
s_idx_sorted.loc[('f', 'a'):('m', 'a')]  #> 튜플 형태의 인덱스로 slicing 가능.
```
