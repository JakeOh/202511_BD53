```python
import numpy as np
import pandas as pd
```

# Series 인덱스

## 단일 계층 인덱스

```python
s = pd.Series(data=np.random.rand(5), name='rand_num')
#> Series 객체도 (데이터프레임의 컬럼처럼) 값들의 이름을 설정할 수 있음.
s
```

```python
s.index  #> Series 객체가 가지고 있는 index 속성
```

```python
s.values  #> Series 객체가 가지고 있는 values 속성 -> 값들을 아이템으로 갖는 1차원 ndarray(배열)
```

```python
s = pd.Series(data=np.random.randint(100, size=5), name='score',
              index=['scott', 'allen', 'king', '홍길동', '오쌤'])
s
```

```python
s.index
```

```python
s.values
```

```python
s.index.nlevels  # 인덱스의 레벨(계층) 개수
```
