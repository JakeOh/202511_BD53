*   **함수(function)**: 기능을 수행하는 코드 블록.
*   **인수(argument)**: 함수를 호출할 때 함수에게 전달하는 값.
*   **매개변수(parameter)**: 아규먼트를 저장하기 위해서 함수를 정의하는 곳에서 선언하는 변수.
*   **반환값(return value)**: 함수가 기능을 수행하고 그 결과로 함수를 호출한 곳에 되돌려 주는 값.
    *   함수에 따라서 반환값이 있는 함수도 있고, 반환값이 없는 함수도 있음.

# 함수 선언(정의)

```
def function_name([parameter1, parameter2, ...]):
    ["""문서 주석(documentation comment)"""]
    함수가 수행할 코드 블록
    [return 반환값]
```

*   반환값이 없는 함수는 return 문장을 생략하거나, `return None` 문장을 사용할 수 있음.
*   return의 의미:
    *   함수 종료.
    *   함수를 호출한 곳에 값을 반환.


````python
# 함수 정의(선언)
def subtract(x, y):
    """숫자 x, y를 전달 받아서 x - y를 반환하는 함수."""
    return x - y
````

````python
result = subtract(2, 1)  # 함수 호출(call, invoke)
print(result)
````

## 값을 반환하지 않는 함수

````python
def repeat_message(msg, n):
    """
    msg를 n번 반복해서 출력.

    @param msg - 출력할 문자열.
    @param n - 반복해서 출력할 횟수. 0 또는 양의 정수.
    """
    for _ in range(n):
        print(msg)

    # return None
````

````python
repeat_message('안녕하세요', 3)
````

````python
repeat_message('Hello, Python!', 5)
````

## 2개 이상의 값을 반환하는 함수

````python
def plus_and_minus(x, y):
    """
    x + y, x - y를 반환.

    @param x - 숫자 타입(int 또는 float).
    @param y - 숫자 타입(int 또는 float).
    @return x + y, x - y
    """
    return x + y, x - y
````

````python
a, b = plus_and_minus(1.0, 2.0)
print(f'a={a}, b={b}')
````

````python
result = plus_and_minus(3, 7)
print(result)  # result의 데이터 타입: tuple
print(result[0])  # tuple의 첫번째 원소
print(result[1])  # tuple의 두번째 원소
````
