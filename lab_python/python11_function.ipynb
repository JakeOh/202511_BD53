*   **함수(function)**: 기능을 수행하는 코드 블록.
*   **인수(argument)**: 함수를 호출할 때 함수에게 전달하는 값.
*   **매개변수(parameter)**: 아규먼트를 저장하기 위해서 함수를 정의하는 곳에서 선언하는 변수.
*   **반환값(return value)**: 함수가 기능을 수행하고 그 결과로 함수를 호출한 곳에 되돌려 주는 값.
    *   함수에 따라서 반환값이 있는 함수도 있고, 반환값이 없는 함수도 있음.

# 함수 선언(정의)

```
def function_name([parameter1, parameter2, ...]):
    ["""문서 주석(documentation comment)"""]
    함수가 수행할 코드 블록
    [return 반환값]
```

*   반환값이 없는 함수는 return 문장을 생략하거나, `return None` 문장을 사용할 수 있음.
*   return의 의미:
    *   함수 종료.
    *   함수를 호출한 곳에 값을 반환.


````python
# 함수 정의(선언)
def subtract(x, y):
    """숫자 x, y를 전달 받아서 x - y를 반환하는 함수."""
    return x - y
````

````python
result = subtract(2, 1)  # 함수 호출(call, invoke)
print(result)
````

## 값을 반환하지 않는 함수

````python
def repeat_message(msg, n):
    """
    msg를 n번 반복해서 출력.

    @param msg - 출력할 문자열.
    @param n - 반복해서 출력할 횟수. 0 또는 양의 정수.
    """
    for _ in range(n):
        print(msg)

    # return None
````

````python
repeat_message('안녕하세요', 3)
````

````python
repeat_message('Hello, Python!', 5)
````

## 2개 이상의 값을 반환하는 함수

````python
def plus_and_minus(x, y):
    """
    x + y, x - y를 반환.

    @param x - 숫자 타입(int 또는 float).
    @param y - 숫자 타입(int 또는 float).
    @return x + y, x - y
    """
    return x + y, x - y
````

````python
a, b = plus_and_minus(1.0, 2.0)
print(f'a={a}, b={b}')
````

````python
result = plus_and_minus(3, 7)
print(result)  # result의 데이터 타입: tuple
print(result[0])  # tuple의 첫번째 원소
print(result[1])  # tuple의 두번째 원소
````

### Ex. 1

*   함수 이름: make_list
*   기능: [start, end) 범위의 정수 난수 n개를 아이템으로 갖는 리스트를 반환.
*   파라미터: start(정수), end(정수), n(0 또는 양의 정수)
*   반환값: 정수 난수들의 리스트.


````python
def make_list(start, end, n):
    """
    start 이상 end 미만의 정수 난수 n개를 아이템으로 갖는 리스트를 리턴.

    @param start - int. 난수 시작 범위.
    @param end - int. 난수 종료 범위.
    @param n - 리스트의 아이템 개수(난수 개수). 0 또는 양의 정수.
    @return 정수 난수 n개를 갖는 리스트.
    """

    # numbers = []
    # for _ in range(n):
    #     numbers.append(random.randrange(start, end))
    numbers = [random.randrange(start, end) for _ in range(n)]

    return numbers
````

````python
rand_numbers = make_list(0, 10, 5)
print(rand_numbers)
````

### Ex. 2

*   함수 이름: calculate_sum
*   기능: 숫자(정수 또는 실수)들을 원소로 갖는 리스트를 전달받아서, 그 리스트의 모든 원소들의 합을 리턴.
*   파라미터: array(숫자들의 리스트).
*   반환값: 리스트 array의 모든 원소들의 합계.


````python
def calculate_sum(array):
    """
    숫자들의 리스트 array의 모든 원소들의 합을 리턴.

    @param array - 숫자들을 원소로 갖는 리스트.
    @return array의 모든 원소들의 합.
    """

    total = 0
    for x in array:
        total += x

    return total
````

````python
calculate_sum([1, 2, 3])
````

````python
calculate_sum(rand_numbers)
````

### Ex. 3

*   함수 이름: calculate_mean
*   기능: 숫자들을 원소로 갖는 리스트의 원소들의 평균을 리턴.
*   파라미터: array(숫자들의 리스트).
*   반환값: 리스트 array의 모든 원소들의 평균.


````python
def calculate_mean(array):
    """
    숫자들의 리스트 array의 원소들의 평균을 리턴.

    @param array - 숫자들을 원소로 갖는 리스트.
    @return array 원소들의 평균.
    """

    mean = calculate_sum(array) / len(array)

    return mean
````

````python
calculate_mean(rand_numbers)
````

### Ex. 4

*   함수 이름: find_min_and_max
*   기능: 숫자들을 원소로 갖는 리스트에서 최솟값과 최댓값을 찾아서 리턴.
*   파라미터: array(숫자들의 리스트).
*   반환값: 리스트 array에서 최솟값, 최댓값.


````python
def find_min_and_max(array):
    """
    숫자들의 리스트 array 원소들 중에서 최솟값과 최댓값을 리턴.

    @param array - 숫자들을 저장하는 리스트.
    @return 최솟값, 최댓값
    """

    min_value = array[0]
    max_value = array[0]
    for x in array:
        if x > max_value:
            max_value = x
        if x < min_value:
            min_value = x

    return min_value, max_value
````

````python
find_min_and_max(rand_numbers)
````

# Default Argument(기본 아규먼트)

*   함수를 선언할 때 파라미터에 설정하는 기본값.
    *   default parameter: 기본값이 설정된 파라미터(매개변수).
*   함수를 호출할 때 default parameter에 아규먼트를 주지 않으면, default argument가 기본값으로 사용됨.
*   함수를 호출할 때 default parameter에 아규먼트를 주면, 기본값 대신 전달한 아규먼트가 사용됨.


````python
print('안녕', 'Hello', '파이썬', 'Python')  # sep=' ' 기본값을 사용
````

````python
print('안녕', 'Hello', '파이썬', 'Python', sep=',')  # sep 파라미터의 기본값을 변경.
````

````python
# default argument를 갖는 함수 정의
def repeat_message(msg, n=1):
    for _ in range(n):
        print(msg)
````

````python
repeat_message('안녕하세요.')
````

````python
repeat_message('안녕하세요', 3)
````

**(주의)** default parameter(기본값이 있는 파라미터)는 기본값을 갖지 않는 파라미터(들) 뒤에서만 선언할 수 있음!

default parameter를 기본값을 갖지 않는 파라미터보다 먼저 선언하면 문법 오류!

````python
def test_success(x, y=0):
    pass
````

````python
def test_failure(x=0, y):
    pass
````

````python
def test(x, y):
    pass
````

````python
def test(x=0, y=0):
    print(f'x = {x}, y = {y}')
````

````python
test()
````

````python
test(1)
````

````python
test(1, 2)  # positional argument 방식
````

````python
test(y=1)  # keyword argument 방식
````

함수를 호출할 때 아규먼트를 전달하는 방법:

*   **positional argument**: 선언된 파라미터 순서대로 아규먼트들을 전달하는 방식.
*   **keyword argument**: `param=value` 형식으로 아규먼트(들)을 전달하는 방식.
    *   keyword argument 방식은 파라미터 순서를 지키지 않아도 됨.
    *   **(주의)** 함수를 호출할 때 positional argument와 keyword argument를 함께 사용하는 경우에는, 반드시 positional argument들을 먼저 써야 함.
    *   keyword argument를 positional argument보다 먼저 쓸 수는 없음!


````python
test(y=1, x=123)  # keyword argument는 파라미터 순서를 지키지 않아도 됨.
````

````python
test(1, y=123)  # positional 먼저 -> keyword 나중에
````

````python
test(x=1, 123)  # keyword 먼저, positional 나중에 --> 문법 오류
````

````python
test(x=1, y=12)
````

# 가변길이 인수(variable-length arguments)

*   함수를 호출할 때 아규먼트를 몇 개를 전달해도 됨.
    *   아규먼트를 한 개도 전달하지 않아도 됨.
*   함수를 선언할 때 파라미터 이름 앞에 `*`를 사용한 파라미터가 가변길이 인수가 됨.
*   함수 내부에서 가변길이 인수는 튜플(tuple)로 취급함.
    *   가변길이 인수는 인덱스를 가짐.
    *   가변길이 인수는 `for-in` 구문에서 사용할 수 있음.
*   함수를 선언할 때 가변길이 인수는 1개만 선언할 수 있음.
*   가변길이 인수는 keyword argument 방식으로 아규먼트들을 전달할 수 없음.
*   가변길이 인수 뒤에서 선언된 파라미터(들)은 keyword argument 방식으로만 아규먼트를 전달할 수 있음.


````python
print()  # arg 0개
print('hello')  # arg 1개
print('hello', '안녕')  # arg 2개
````

````python
print('qwer', 'hello', sep='::')
````

````python
# 가변길이 인수를 2개 이상 선언하면 문법 오류
def fn(*a, *b):
    pass
````
