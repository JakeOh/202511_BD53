Colab 코드 셀 실행 방법:


*   실행 버튼 클릭
*   런타임 메뉴에서 실행과 관련된 항목을 선택
*   단축키
    *   Ctrl+Enter: 현재 포커스(초점)이 있는 셀을 실행
    *   Shift+Enter: 현재 포커스가 있는 셀을 실행하고, 그 다음 셀로 포커스를 이동. 만약 다음 셀이 없으면 새로운 코드 셀을 자동으로 만들어 줌.
    *   Ctrl+F9: 모두 실행. Jupyter 노트북의 모든 코드 셀들을 처음부터 순서대로 실행.
    *   Ctrl+F10: 셀 및 하위 셀 실행. 현재 포커스가 있는 셀부터 노트북 마지막의 코드 셀까지 순서대로 실행.
    *   Ctrl+F8: 이전 셀 실행. 현재 포커스가 있는 셀의 이전 코드 셀들을 처음부터 순서대로 실행.

# 여러가지 print 방법

```python
# 인라인 주석(inline comment)
# 파이썬 문장은 문장 끝에 세미콜론(;)을 사용하지 않음.
# 파이썬은 문자열(string)은 작은따옴표('')와 큰따옴표("")를 모두 사용 가능.
# 파이썬의 식별자(예약어, 함수이름)들은 대/소문자를 구분.

print('안녕하세요!!')
print("Hello!!")
print("I'm a boy.")
print('I\'m a boy.')
```
