# list

*   `list`: 여러 개의 값들을 하나의 변수에 저장하기 위한 데이터 타입.
*   아이템(item). 원소(element): 리스트에 저장된 각각의 값.
*   인덱스(index): 리스트에 아이템들이 저장된 위치를 표시하는 정수.
    *   리스트의 인덱스는 **0**부터 시작.
    *   리스트의 마지막 인덱스는 (아이템 개수 - 1)
    *   리스트의 끝에서부터 역순으로 인덱스를 매길 때는 *음수 인덱스*를 사용할 수 있음.
*   리스트는 아이템들을 추가, 변경, 삭제할 수 있음.
    *   `tuple`과 다른 점.


```python
# 숫자들을 저장하는 list
numbers = [1, 2, 10, 20, -1, -100]
print(type(numbers))
print(numbers)
```

```python
# 파이썬의 리스트는 여러가지 타입의 값들을 하나의 변수에 저장할 수 있음.
values = [1, 2, '안녕', '파이썬', 3.141592, True, [1, 2, 3]]
print(type(values))
print(values)
```

# Indexing

`list_name[index]`.
인덱스를 사용해서 리스트의 아이템을 사용하는 방법.

```python
numbers = [1, -2, -3, 4, 5, 100]
numbers
```

```python
numbers[0]  # 리스트 numbers의 인덱스 0 위치(첫번째)의 원소
```

```python
numbers[1]  # 인덱스 1번 위치(두번째)의 원소.
```

```python
numbers[5]
```

```python
numbers[-1]  # 끝에서 첫번째 아이템. 마지막 아이템.
```

```python
numbers[-2]  # 끝에서 두번째 아이템.
```

```python
print(numbers[-1])
print(numbers[-2])
```

# Slicing

`list_name[start:end]`

*   리스트에서 start <= index < end 범위의 아이템들을 잘라내는 방법.
*   시작 인덱스 start를 생략하면 첫번째 아이템부터 잘라냄. start의 기본값은 0.
*   종료 인덱스 end를 생략하면 마지막 아이템까지 잘라냄.
*   slicing의 결과는 *새로운 리스트*.


```python
numbers
```

```python
numbers[1:4]
```

```python
numbers[3:6]
```

```python
numbers[3:]  # 인덱스 3번 위치부터 끝까지 자름.
```

```python
numbers[:3]  # 리스트에서 첫 3개 원소를 자름.
```

```python
numbers[-2:]  # 리스트에서 마지막 2개 원소를 자름.
```

```python
numbers[:1]  # slicing: 리스트에서 첫번째 원소만 잘라낸 새로운 "리스트"
```

```python
numbers[0]  # indexing: 해당 인덱스 위치의 원소를 읽어옴.
```

## 문자열(string)

*   문자들의 리스트.
    *   indexing
    *   slicing

```python
message = '안녕, Python!'
```

```python
message[0]  # 문자열에서 첫번째 글자
```

```python
message[-1]  # 문자열에서 마지막 글자
```

```python
message[:3]  # 문자열에서 첫 3글자를 선택
```

```python
message[-3:]  # 문자열에서 마지막 3글자
```

# 리스트의 산술 연산

*   `list + list`: concatenate. 2개의 리스트의 아이템들을 합쳐서 하나의 리스트로 만듦.
*   `list * int`, `int * list`: replicate. 리스트의 아이템들을 정수만큼 복사해서 아이템 개수를 늘림.


```python
print('안녕' + 'Python')
print('안녕' * 3)
```

```python
[1, 2, 3] + [10, 20, 30]
```

```python
[1, 2, 3] * 3
```
