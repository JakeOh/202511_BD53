# list

*   `list`: 여러 개의 값들을 하나의 변수에 저장하기 위한 데이터 타입.
*   아이템(item). 원소(element): 리스트에 저장된 각각의 값.
*   인덱스(index): 리스트에 아이템들이 저장된 위치를 표시하는 정수.
    *   리스트의 인덱스는 **0**부터 시작.
    *   리스트의 마지막 인덱스는 (아이템 개수 - 1)
    *   리스트의 끝에서부터 역순으로 인덱스를 매길 때는 *음수 인덱스*를 사용할 수 있음.
*   리스트는 아이템들을 추가, 변경, 삭제할 수 있음.
    *   `tuple`과 다른 점.


````python
# 숫자들을 저장하는 list
numbers = [1, 2, 10, 20, -1, -100]
print(type(numbers))
print(numbers)
````

````python
# 파이썬의 리스트는 여러가지 타입의 값들을 하나의 변수에 저장할 수 있음.
values = [1, 2, '안녕', '파이썬', 3.141592, True, [1, 2, 3]]
print(type(values))
print(values)
````

# Indexing

````python
numbers = [1, -2, -3, 4, 5, 100]
numbers
````

````python
numbers[0]  # 리스트 numbers의 인덱스 0 위치(첫번째)의 원소
````

````python
numbers[1]  # 인덱스 1번 위치(두번째)의 원소.
````

````python
numbers[5]
````

````python
numbers[-1]  # 끝에서 첫번째 아이템. 마지막 아이템.
````

````python
numbers[-2]  # 끝에서 두번째 아이템.
````

````python
print(numbers[-1])
print(numbers[-2])
````

# Slicing

`list_name[start:end]`

*   리스트에서 start <= index < end 범위의 아이템들을 잘라내는 방법.
*   시작 인덱스 start를 생략하면 첫번째 아이템부터 잘라냄. start의 기본값은 0.
*   종료 인덱스 end를 생략하면 마지막 아이템까지 잘라냄.
*   slicing의 결과는 *새로운 리스트*.


````python
numbers
````

````python
numbers[1:4]
````

````python
numbers[3:6]
````

````python
numbers[3:]  # 인덱스 3번 위치부터 끝까지 자름.
````

````python
numbers[:3]  # 리스트에서 첫 3개 원소를 자름.
````

````python
numbers[-2:]  # 리스트에서 마지막 2개 원소를 자름.
````

````python
numbers[:1]  # slicing: 리스트에서 첫번째 원소만 잘라낸 새로운 "리스트"
````

````python
numbers[0]  # indexing: 해당 인덱스 위치의 원소를 읽어옴.
````

# 리스트의 산술 연산

*   `list + list`: concatenate. 2개의 리스트의 아이템들을 합쳐서 하나의 리스트로 만듦.
*   `list * int`, `int * list`: replicate. 리스트의 아이템들을 정수만큼 복사해서 아이템 개수를 늘림.

# 리스트의 기능들 - 메서드(함수)

## 원소 추가

````python
numbers = [1, 2, 3, 4, 5]
numbers
````

````python
# list.append(val): list의 끝에 val을 아이템으로 추가.
numbers.append(100)
numbers
````

````python
# list.insert(idx, val): list의 인덱스 idx 앞에 값 val을 삽입(추가).
numbers.insert(0, -1)
numbers
````

````python
numbers.insert(2, -2)
numbers
````

````python
# list.extend(iterable): iterable 안에 있는 아이템들을 "하나씩" 리스트의 끝에 추가.
numbers.extend([11, 12, 13])
numbers
````

````python
# extend() 메서드 기능은 + 연산 결과와 같다.
numbers = numbers + [21, 22, 23]
numbers
````

````python
# (주의) append와 extend의 차이점!
numbers.append([101, 102, 103])
numbers
````

## 원소 삭제

````python
numbers = [1, 2, 3, 10, 20, 30, 5, 6, 7]
numbers
````

````python
# list.pop(): list의 가장 마지막 아이템을 삭제.
numbers.pop()
numbers
````

````python
# list.pop(index): list에서 index 위치의 아이템을 삭제.
numbers.pop(0)  # 인덱스 0 위치의 아이템(첫번째 아이템)을 삭제.
numbers
````

````python
# list.remove(value): list에서 첫번째로 등장하는 값 value를 삭제.
numbers.remove(2)
numbers
````

````python
test = [1, 2, 3] * 3
print(test)  # 삭제 전
test.remove(2)
print(test)  # 삭제 후
````

````python
# list.clear(): list의 모든 아이템들을 삭제.
numbers.clear()
numbers
````

## 원소 개수 세기

````python
numbers = [1, 2, 3, 4, 5, 6, 1, 2, 1, 2, 3]
numbers
````

````python
# list.count(value): list에서 value가 몇 번 등장하는 지 개수를 반환.
numbers.count(3)
````

````python
# len(iterable) 함수: iterable의 아이템 개수를 반환.
len(numbers)  # 리스트 numbers의 아이템 개수
````

````python
len('Hello')  # 문자열 'Hello'의 문자 개수.
````

# list와 for 반복문

````python
languages = ['SQL', 'Python', 'R']
for x in languages:
    print(x)
````

````python
# list에서 인덱스와 값을 같이 꺼내서 반복(iteration)을 하고 싶을 때.
for index, value in enumerate(languages):
    print(index, ':', value)
````
