## 파생변수 추가:

*   mpg 데이터프레임에서 name 컬럼의 첫번째 단어만 선택해서 company 컬럼을 추가.


```python
# 데이터프레임에 파생변수 추가: dataframe_name['new_column'] = value
mpg['company'] = mpg.name.str.split().str[0]
```

## 데이터 정제

자동차 회사 이름의 오타 수정:

*   chevroelt --> chevrolet
*   maxda --> mazda
*   mercedes --> mercedes-benz
*   toyouta --> toyota
*   vokswagen --> volkswagen
*   vw --> volkswagen


```python
mpg.loc[mpg.company == 'chevroelt', 'company'] = 'chevrolet'
mpg.loc[mpg.company == 'maxda', 'company'] = 'mazda'
mpg.loc[mpg.company == 'mercedes', 'company'] = 'mercedes-benz'
mpg.loc[mpg.company == 'toyouta', 'company'] = 'toyota'
mpg.loc[mpg.company == 'vokswagen', 'company'] = 'volkswagen'
mpg.loc[mpg.company == 'vw', 'company'] = 'volkswagen'
```

```python
companies = mpg.company.unique()
companies
```

```python
companies.size
```

회사 이름의 빈도수

```python
company_count = mpg.company.value_counts()
company_count
```

```python
sns.countplot(data=mpg, y='company')
plt.show()
```

```python
plt.barh(y=company_count.index, width=company_count.values)
plt.show()
```

```python
company_count2 = mpg.company.value_counts().sort_index()
company_count2
```

```python
# plt.figure(figsize=(8, 12))
plt.barh(y=company_count2.index, width=company_count2.values)
plt.show()
```

회사별 연비 평균 상위/하위 10개 회사

```python
mpg_by_company = mpg.groupby(by=['company']).mpg.mean()
mpg_by_company
```

```python
# 연비 평균 상위 10개 회사
top10 = mpg_by_company.sort_values(ascending=False).head(n=10)
top10
```

```python
plt.barh(y=top10.index, width=top10.values)
plt.xlabel('mean of mpg')
plt.show()
```

```python
bottom10 = mpg_by_company.sort_values().head(n=10)
bottom10
```

```python
plt.barh(y=bottom10.index, width=bottom10.values)
plt.show()
```

# movies 데이터프레임

```python
movies_file = 'https://github.com/wesm/pydata-book/raw/refs/heads/3rd-edition/datasets/movielens/movies.dat'
```

movies.dat 파일은 값들이 '::'으로 구분된 텍스트 파일. 파일에는 컬럼 이름이 없음(첫번째 줄부터 모두 데이터).

`read_csv()` 함수를 사용해서 DataFrame을 생성할 수 있음.

```python
movies = pd.read_csv(movies_file,
                     sep='::',                              # sep: 구분자
                     names=['movie_id', 'title', 'genres'], # names: 컬럼 이름
                     engine='python')                       # engine='python': 구분자가 한 글자 이상인 경우.
```

```python
movies.head()
```

```python
movies.tail()
```

```python
movies.info()
```

*   개봉 연도 분석 - 연도별 개봉 영화 숫자
    *   title 컬럼의 문자열에서 개봉연도를 추출해서 파생변수(year)를 추가.
*   장르 분석
    *   장르가 Animation인 영화는 몇 개?
    *   가장 많이 등장하는 장르 이름?
        *   가장 많이 등장하는 장르 이름 상위 10개

## 영화 개봉 연도 분석

```python
# year 파생변수 추가
movies['year'] = movies.title.str[-5:-1].astype('int')
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies.year.describe()
```

```python
# 연도별 개봉 영화 수
year_count = movies.year.value_counts()
year_count
```

```python
sns.barplot(data=year_count)
plt.show()
```

```python
sns.lineplot(data=year_count)
plt.show()
```

## 영화 장르 분석

### 장르가 Animation인 영화

```python
animations = movies[movies.genres.str.contains('Animation')]
```

```python
animations
```

```python
len(animations)  # 데이프레임의 행의 개수
```

```python
# 뮤지컬 애니메이션
musical_animations = movies[movies.genres.str.contains('Animation') &
                            movies.genres.str.contains('Musical')]
```

```python
musical_animations.head(n=8)
```

```python
len(musical_animations)
```

```python
# 로맨틱 코미디 장르(Romance & Comedy)의 영화들
romantic_comedies = movies[movies.genres.str.contains('Romance') &
                           movies.genres.str.contains('Comedy')]
```

```python
romantic_comedies.head(n=8)
```

```python
len(romantic_comedies)
```
