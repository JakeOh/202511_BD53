# tuple

*   tuple(튜플): 저장된 아이템들을 변경할 수 없는 리스트.
*   인덱스 기반의 데이터 타입
    *   indexing: `tuple_name[index]`. 특정 인덱스 위치의 아이템을 읽어옴.
    *   slicing: `tuple_name[start:end]`. 특정 인덱스 범위의 아이템들로 이루어진 새로운 튜플을 만듦.
*   append, extend, insert, pop, remove, ... 원소를 추가/삭제/변경하는 기능들은 제공되지 않음.


```python
numbers = (1, 2, 3, 1, 1, 2)
print(type(numbers))
print(numbers)
```

```python
numbers.count(1)  # tuple|list.count(value): 튜플 또는 리스트에서 value가 몇 개 있는 지 반환.
```

```python
numbers.index(1)  # tuple|list.index(value): 튜플 또는 리스트에서 value가 처음 나오는 인덱스를 반환.
```

```python
len(numbers)  # len(tuple|list): 튜플 또는 리스트의 원소 개수를 반환.
```

# Indexing

```python
numbers[0]  # 튜플의 첫번째 아이템
```

```python
numbers[-1]  # 튜플의 마지막 아이템
```

# Slicing

```python
numbers[:3]  # 튜플의 첫 3개 원소
```

```python
numbers[-3:]  # 튜플에서 마지막 3개 원소
```

# tuple과 for 반복문

```python
numbers = (1, 10, -2, -20, 30)
for x in numbers:
    print(x)
```

```python
for index, value in enumerate(numbers):
    print(index, ':', value)
```

```python
numbers1 = (1, 2, 3)
numbers2 = (10, 20, 30)
for x, y in zip(numbers1, numbers2):
    print(x, y)
```

## 2차원 튜플과 for 반복문

```python
for x in matrix:
    print(x)
```

```python
for row in matrix:
    for item in row:
        print(item, end='\t')
    print()
```
