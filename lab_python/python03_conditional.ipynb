# 조건문(conditional statement)

## else 없는 조건문

```
if 조건식:
    조건식을 만족할 때(조건식이 참일 때) 실행할 코드 블록(문장들)
```

*   `if 조건식` 뒤에는 반드시 콜론(:)을 써야 됨.
*   if 블록은 반드시 들여쓰기(indentation)을 해야 됨.
*   들여쓰기 크기는 일정해야 함.


````python
age = int(input('나이 입력>> '))
if age >= 0:
    print('안녕하세요...')
    print(f'당신은 {age}살이군요...')
````

## else 있는 조건문

```
if 조건식:
    조건식을 만족할 때 실행할 코드 블록
else:
    조건식을 만족하지 못할 때 실행할 코드 블록
```


````python
age = int(input('나이 입력>> '))
if age >= 0:
    print(f'당신은 {age}살')
else:
    print('나이는 음수가 될 수 없어요.')
````

(연습 문제) 정수를 입력받아서, 그 정수가 짝수인 지 홀수인 지를 출력하세요.

````python
number = int(input('정수 입력>> '))
if number % 2 == 0:  # number를 2로 나눈 나머지가 0이면
    print('짝수')
else:  # 그렇지 않으면
    print('홀수')
````

Falsy values: 파이썬에서 False로 취급하는 값들.

*   숫자(int, float): 0
*   문자열(str): '' (빈 문자열, empty string)
*   빈 리스트(empty list)
*   None: 변수에 값이 할당되어 있지 않은 경우(값이 없는 경우)를 표현하는 파이썬 예약어(키워드).


````python
number = int(input('정수 입력>> '))
if number % 2:  # number를 2로 나눈 나머지가 있으면
    print('홀수')
else:  # 그렇지 않으면(2로 나눈 나머지가 없으면)
    print('짝수')
````
