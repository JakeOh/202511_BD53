# Python Data Type(자료 타입)

*   숫자 타입
    *   정수(int): integer
    *   실수(float): floating-point number
*   문자열 타입(str): string
*   논리 타입(bool): True/False


## 숫자 타입

```python
# 변수 선언, 값 할당: 변수이름 = 값
n = 123
print(type(n))  # type(변수이름): 변수의 타입을 반환하는 함수.
print('n =', n)
```

```python
n = 3.141592
print(type(n))
print('n =', n)
```

## 문자열 타입

파이썬의 문자열은 작은따옴표('')와 큰따옴표("")를 모두 사용할 수 있음.

```python
name = '오쌤'
print(type(name))
print('name =', name)
```
