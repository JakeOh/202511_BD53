# dict

*   사전(dictionary) 형식의 데이터 타입.
*   키(key)와 값(value)의 쌍으로 저장하는 데이터 타입.
    *   list, tuple: 인덱스를 기반으로 값들을 저장하는 데이터 타입.
    *   dict: 키를 기반으로 값을 저장하는 데이터 타입.
    *   dict 타입에서 키의 역할은 list나 tuple에서 인덱스의 역할과 비슷.
        *   dict에 저장된 1개의 값을 참조할 때 사용.
        *   키는 중복된 값을 허용하지 않음.

```python
students = {100: '홍길동', 101: '오쌤', 200: 'Scott', 201: '홍길동'}
print(type(students))
print(students)
```
