# dict

*   사전(dictionary) 형식의 데이터 타입.
*   키(key)와 값(value)의 쌍으로 저장하는 데이터 타입.
    *   list, tuple: 인덱스를 기반으로 값들을 저장하는 데이터 타입.
    *   dict: 키를 기반으로 값을 저장하는 데이터 타입.
    *   dict 타입에서 키의 역할은 list나 tuple에서 인덱스의 역할과 비슷.
        *   dict에 저장된 1개의 값을 참조할 때 사용.
        *   키는 중복된 값을 허용하지 않음.

```python
students = {100: '홍길동', 101: '오쌤', 200: 'Scott', 201: '홍길동'}
print(type(students))
print(students)
```

```python
# dict_name[key] -> value를 반환
students[100]
```

```python
# students[300]
#-> dict에 존재하지 않는 키를 사용하면 KeyError가 발생
```

# dict 메서드

```python
# dict_name.keys(): dict가 가지고 있는 키들의 리스트.
students.keys()
```

```python
# dict_name.values(): dict가 가지고 있는 값들의 리스트.
students.values()
```

```python
# dict_name.items(): (key, value) 튜플들의 리스트.
students.items()
```

```python
# dict_name.get(key): key에 해당하는 값(value)를 리턴(반환).
# dict_name[key]과 비슷. get() 메서드는 존재하지 않는 키일 때 에러가 발생하지 않음.
students.get(100)
```

`dict_name[key] = value`

*   key가 dict에 있을 때, key에 매핑된 값을 새로운 값으로 변경.
*   key가 dict에 없을 때, key-value 아이템을 dict에 추가.


```python
print(students)
students[201] = '김길동'  # 키가 존재할 때
print(students)
```

```python
students[300] = 'King'  # 키가 존재하지 않을 때
print(students)
```

```python
students.items()
```

# dict와 for 반복문

```python
# for-in 반복문에서 dict를 사용하면 key들을 순서대로 반복(iteration).
for k in students:
    print(k, ':', students[k])  # students.get(k)
```

```python
for k, v in students.items():
    print(k, ':', v)
```

```python
book = {'title': '파이썬 예제', 'author': '오쌤', 'price': 10_000}
print(book)
print(book['title'])
print(book.get('author'))
```

```python
for k, v in book.items():
    print(k, ':', v)
```

# 다양한 타입의 자료들을 저장하는 dict

```python
contact = {
    'id': 1,
    'name': '오쌤',
    'phone': ['010-0000-0000', '02-0000-0000', '02-1234-5678'],
    'email': {'company': 'jake@itwill.com', 'personal': 'jake@gmail.com'}
}
print(contact)
```

```python
contact['id']  # contact.get('id')
```

```python
# 첫번째 전화번호
contact['phone'][0]
```

```python
contact.get('phone')[0]
```

```python
# 회사 이메일
contact['email']['company']
```

```python
contact.get('email').get('company')
```
