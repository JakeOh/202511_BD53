# `str` 클래스 메서드

```python
s = 'heLLo, pyTHon! hello PYTHON. 안녕하세요?'
```

```python
type(s)  # type(arg): arg의 타입(클래스 이름)을 리턴.
```

```python
print(s)  # 변수 s의 __str__ 메서드가 리턴하는 문자열을 출력.
```

```python
s  # 변수 s의 __repr__ 메서드가 리턴하는 문자열을 출력.
```

```python
result = s.lower()  # 문자열 s를 소문자로 변환한 "새로운" 문자열을 리턴. s는 변경되지 않음!
print(result)
print(s)
```

```python
result = s.upper()  # 문자열 s를 대문자로 변환한 "새로운" 문자열 리턴.
print(result)
print(s)
```

```python
s.capitalize()  # 문자열의 첫글자만 대문자로, 나머지는 소문자로 변환한 문자열을 리턴.
```

```python
s.count('h')  # 문자열 s에서 'h'가 몇 번 나오는 지를 리턴.
```

```python
# 대/소문자 구분없이 h가 몇 개 있는 지?
s.lower().count('h')
```

```python
s.upper().count('H')
```

```python
email = 'gildong@itwill.com'

idx = email.find('@')  # 문자열 email에서 '@'가 있는 인덱스를 리턴.
print(idx)

# 문자열 email에서 아이디(gildong)와 도메인(itwill.com)을 구분해서 출력
print(email[:idx])  # 문자열 slicing
print(email[(idx + 1):])
```

```python
msg = '파이썬 쉽죠? 열심히 공부합시다!'
result = msg.split()
#> 문자열 msg를 공백(whitespace: space, tab, new line)을 기준으로 자른 문자열들의 리스트를 리턴.
print(result)
```

```python
date_str = '2025-12-17'
result = date_str.split('-')
#> 문자열 date_str을 '-'를 기준으로 자른 문자열들의 리스트를 리턴.
print(result)
```

```python
print(email)
result = email.split('@')
print(result)
print(result[0])
print(result[1])
```

```python
# csv(comma-separated values)
csv_str = '1,홍길동,16,학생'
csv_str.split(',')
```

```python
'/'.join(['2025', '12', '17'])
#> 아규먼트로 전달된 리스트를 '/'를 구분자로 하는 문자열로 변환해서 리턴.
```

```python
':'.join(['a', 'b', 'c', 'd'])  #> 리스트를 ':'를 구분자로 하는 문자열로 변환해서 리턴.
```

```python
sentence = '        Hello, Python!         안녕하세요.        '
print(sentence)
print(sentence.lstrip())
print(sentence.rstrip())
print(sentence.strip())
```
