# Categorical Data 다루기

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(data={
    'user_id': np.arange(1, 11),
    'gender': ['M'] * 5 + ['F'] * 5,
    'score': np.random.randint(1, 6, size=10)
})
```

```python
df
```

```python
df.info()
```

```python
# gender 컬럼을 문자열 타입에서 카테고리 타입으로 변환.
df.gender = df.gender.astype('category')
```

```python
df
```

```python
df.info()
```

```python
# Categorical Accessor(카테고리 접근자)
df.gender.cat.categories
```

# 연속형 변수에서 파생된 카테고리 변수 만들기

```python
df = pd.DataFrame(data={
    'id': np.arange(1, 11),
    'age': np.random.randint(100, size=10)
})
```

```python
df
```

```python
# age 컬럼의 값을 이용해서 연령대 파생변수를 추가.
df['age_range'] = pd.cut(x=df.age, bins=np.arange(0, 110, 10), right=False)
```

```python
df
```

```python
df.info()
```

```python
df.age_range.cat.categories
```

```python
# age 컬럼의 값을 사용한 파생변수
df['age_cat'] = pd.cut(x=df.age, bins=[0, 19, 45, 60, 100], right=False,
                       labels=['미성년', '청년', '장년', '노년'])
```

```python
df
```

```python
df.info()
```

```python
df.age_cat.cat.categories
```
