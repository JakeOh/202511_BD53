# while 반복문

```
[초기화 문장]
while 조건식:
    조건식을 만족하는 동안에 반복할 코드 블록
    [조건을 변경할 수 있는 문장]
```

*   조건식을 만족하는 동안에 블록 내부를 반복해서 실행하게 됨.
*   조건식을 만족하지 못하면(조건식이 False가 되면) 반복을 멈춤.
*   `while 조건식` 다음에는 반드시 콜론(:)을 사용해야 됨.
*   블록 안에서는 들여쓰기 크기를 일정하게 작성.


````python
# 1 ~ 5까지 순서대로 출력
n = 1  # 변수 초기화 - 변수를 선언하고, 최초로 값을 할당(저장)하는 것.
while n <= 5:
    print(n)
    n += 1
````

````python
# 5 ~ 1까지 내림차순으로 출력
n = 5
while n > 0:  # while n >= 1: 과 같음.
    print(n)
    n -= 1
````

````python
# 2 4 6 8 10 순서로 출력
n = 2
while n <= 10:
    print(n)
    n += 2
````

````python
n = 2
while n <= 10:
    if n % 2 == 0:
        print(n)
    n += 1
````

````python
# 9 7 5 3 1 순서로 출력
n = 9
while n > 0:
    print(n)
    n -= 2
````

````python
n = 9
while n > 0:
    # if n % 2 == 1:  # n을 2로 나눈 나머지가 1과 같으면
    if n % 2:  # n을 2로 나눈 나머지가 있으면
        print(n)
    n -= 1
````

# `break`와 `continue`

*   `break`: 반복(iteration)을 종료. break가 포함된 가장 가까운 반복문만 빠져나감.
*   `continue`: 다음 반복(iteration)을 계속 수행. continue 아래의 문장들은 건너뛰고, 그 다음 반복을 수행.


````python
n = 1
while n < 6:
    if n == 3:
        break
    print(n)
    n += 1
````

````python
n = 0
while n < 5:
    n += 1
    if n == 3:
        continue
    print(n)
````
