# while 반복문

```
[초기화 문장]
while 조건식:
    조건식을 만족하는 동안에 반복할 코드 블록
    [조건을 변경할 수 있는 문장]
```

*   조건식을 만족하는 동안에 블록 내부를 반복해서 실행하게 됨.
*   조건식을 만족하지 못하면(조건식이 False가 되면) 반복을 멈춤.
*   `while 조건식` 다음에는 반드시 콜론(:)을 사용해야 됨.
*   블록 안에서는 들여쓰기 크기를 일정하게 작성.


````python
# 1 ~ 5까지 순서대로 출력
n = 1  # 변수 초기화 - 변수를 선언하고, 최초로 값을 할당(저장)하는 것.
while n <= 5:
    print(n)
    n += 1
````

````python
# 5 ~ 1까지 내림차순으로 출력
n = 5
while n > 0:  # while n >= 1: 과 같음.
    print(n)
    n -= 1
````

````python
# 2 4 6 8 10 순서로 출력
n = 2
while n <= 10:
    print(n)
    n += 2
````

````python
n = 2
while n <= 10:
    if n % 2 == 0:
        print(n)
    n += 1
````
