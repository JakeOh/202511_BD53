# for 반복문

```
for 변수 in iterable_type:
    반복해서 실행할 코드 블록
```

*   iterable_type: 반복(iteration)이 가능한 데이터 타입.
    *   str, range, list, tuple, dict, ...
*   for 반복문은 iterable type에서 아이템들을 순서대로 하나씩 꺼내서 변수에 저장하고 코드 블록 실행.
*   for 반복문은 iterable type에서 더이상 꺼낼 아이템을 없을 때 종료됨.


````python
for s in '안녕하세요':
    print(s)
````

# for 반복문과 `range()` 함수

````python
# range(start, end): start가 end보다 작은 경우, [start, end) 범위의 1씩 증가하는 정수들을 만듦.
for x in range(1, 6):
    print(x)
````

````python
# range(end): [0, end) 범위의 1씩 증가하는 정수들을 만듦.
for x in range(6):
    print(x)
````

````python
# range(start, end, step): start가 end보다 작고 step은 양의 정수인 경우,
# [start, end) 범위의 step 만큼씩 증가하는 정수들을 만듦.
for x in range(0, 10, 2):
    print(x)
````

````python
for x in range(1, 10, 2):
    print(x)
````

````python
# range(start, end, step): start가 end보다 크고 step은 음의 정수인 경우,
# start >= x > end 인 범위의 step 만큼씩 감소하는 정수들을 만듦.
for x in range(5, 0, -1):
    print(x)
````

````python
for x in range(1, 6):
    print('*' * x)
````

````python
for x in range(5, 0, -1):
    print('*' * x)
````

# break vs continue

*   `break`: iteration을 멈춤.
*   `continue`: 그 다음 iteration을 수행.


````python
for x in range(5):
    if x == 2:
        break
    print(x)
````

````python
for x in range(5):
    if x == 2:
        continue
    print(x)
````

for-in 반복문과 range() 함수를 사용해서 구구단 2단은 2x2까지, 3단은 3x3까지, ..., 9단은 9x9까지 출력.

````python
for x in range(2, 10):
    print(f'=== {x} 단 ===')
    for y in range(1, 10):
        print(f'{x} x {y} = {x * y}')
        if x == y:
            break  # for y in range() 반복을 멈춤.
````

````python
for x in range(2, 10):
    print(f'=== {x} 단 ===')
    for y in range(1, x + 1):
        print(f'{x} x {y} = {x * y}')
````

# `print()` 함수

*   탈출 문자열(escape string):
    *   `'\n'`: 줄바꿈(new line)
    *   `'\t'`: 탭(tab)
    *   `'\''`: 작은따옴표. 문자열 안에 인용부호를 포함하고 싶을 때 사용.
*   `print()` 함수의 파라미터(parameter)
    *   `end`: print 함수가 아규먼트들을 출력한 후 마지막에 어떤 문자열을 추가할 것인 지를 설정. 기본값은 줄바꿈(`\n`).
    *   `sep`: separator(구분자). print 함수가 아규먼트들을 출력할 때, 각각의 아규먼트들을 구분하기 위한 문자열을 설정. 기본값이 공백(' ') 1개.


````python
print('I\'m a boy.')
````

````python
print('안녕하세요.')
print('월요일 아침입니다.')
````

````python
print('안녕하세요.', end='::')
print('월요일 아침입니다.')
print('Hello!')
````

````python
print('안녕', 'Hello')
````

````python
print('안녕', 'Hello', sep=',')
````
