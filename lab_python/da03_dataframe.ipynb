# 파생변수

```python
tips.head()
```

```python
# 영수증 금액 대비 팁의 비율
tips.tip / tips.total_bill
```

```python
# 파생변수 추가
tips['tip_ratio'] = tips.tip / tips.total_bill
```

```python
tips.head()
```

*   팁/영수증 비율의 평균값
*   팁/영수증 비율이 평균보다 큰 행들
*   팁/영수증 비율이 평균보다 큰 데이터에서 남성/여성의 비율

```python
tips.tip_ratio.mean()
```

```python
tips.tip_ratio.describe()
```

```python
df = tips[tips.tip_ratio > tips.tip_ratio.mean()]
```

```python
df
```

```python
df.sex.value_counts()
```

```python
# 전체 남성/여성 고객들 중에서 팁-영수증 비율이 평균보다 큰 남성/여성 고객 비율
df.sex.value_counts() / tips.sex.value_counts()
```

# titanic 데이터셋

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

titanic 데이터프레임의 컬럼들:

*   survived: 생존 여부(0=사망, 1=생존)
    *   alive 컬럼과 같은 내용(no=사망, yes=생존)
*   pclass: 객실 등급
    *   class 컬럼과 같은 내용(1=First, 2=Second, 3=Third)
*   sex: 성별(male/female)
*   who: 성인남성(man), 성인여성(woman), 어린이(child)
*   adult_male: 성인 남성 여부(True, False)
*   age: 나이
*   sibsp: 동승한 형제자매(siblings) 또는 배우자(spouses) 수
*   parch: 동승한 부모(parents) 또는 자녀(children) 수
*   alone: 동승자 여부(True=동승자없음(혼자), False=동승자있음(혼자가 아님))
*   fare: 운임(탑승권 가격)
