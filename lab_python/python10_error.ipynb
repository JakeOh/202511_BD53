*   문법적인 오류가 아니라 예외적인 상황이 발생할 수 있는 코드인 경우에는 예외 처리를 해야 함.


# try-except 구문

```
try:
    (1) 실행할 코드 블록
except [에러 이름 [as 별명]]:
    (2) 예외 상황이 발생했을 때 예외를 처리하는 코드 블록
[
else:
    (3) try 블록의 모든 코드가 예외 없이 정상 수행됐을 때 실행할 코드 블록
finally:
    (4) 예외 발생 여부와 상관없이 항상 실행할 코드 블록
]
```

*   try 구문의 실행 순서
    *   예외가 발생한 경우: (1) -> (2) -> (4)
    *   예외가 발생하지 않은 경우: (1) -> (3) -> (4)
*   except 절에서 에러 이름을 생략하면 모든 종류의 에러들을 처리할 수 있음.
*   except 절에서 에러 이름을 지정하면, 선언된 에러만 처리할 수 있음.
*   처리해야 할 에러의 종류가 여러 개인 경우에는 except 절을 여러 개 사용할 수도 있음.


````python
try:
    x = int(input('정수 x = '))
    y = int(input('정수 y = '))
    print(f'{x} + {y} = {x + y}')
    print(f'{x} - {y} = {x - y}')
    print(f'{x} x {y} = {x * y}')
    print(f'{x} / {y} = {x / y}')
except ValueError as e:
    print('x와 y는 정수로 입력해야 합니다.')
    print(e)
except ZeroDivisionError as e:
    print('0으로 나눌 수는 없습니다.')
    print(e)
else:
    print('사칙 연산을 모두 수행했습니다.')
finally:
    print('코드 셀 종료')
````

사용자에게 숫자(실수, float)를 입력하는 프롬프트를 보여주고,

*   사용자가 입력한 문자열이 숫자로 변환될 수 있는 경우에는, 그 숫자를 출력하고 종료.
*   사용자가 입력한 문자열이 숫자로 변환될 수 없는 경우에는, 숫자를 입력하라는 프롬트프를 다시 보여줌.


````python
while True:
    try:
        number = float(input('숫자 입력>> '))
        print(f'number = {number}')
        break
    except:
        print('입력값은 숫자여야 합니다.')
````
